```python
# Importar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('unknown.jpeg')
print("el tama√±o de la imagen es: ", bgr.shape)

#Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

#Mostrar la imagen
plt.imshow(rgb)
```

```python
#Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Mostrar las compenentes de color de la imagen
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Componente Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Componente Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Componente Roja')
plt.show()
```

```python
#Mostrar las componentes HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Componente H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Componente L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Componente S')
plt.show()
```
